```python
%matplotlib inline
import numpy as np
import math
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
```

```python
def edges_full(i):
  n = i * ( i - 1 ) / 2
  return(n)
```

```python
def r_nodes(n):
  num = 0
  for i in range(1, n+1):
    num += 2**i
  return(num)  

def r_distance(i):
  if i == 2:
    n = 1
  else:
    n = 2 * math.log2(i+2) - 4
  return n

def r_num_edges(i):
  n = 2 * i - 2
  return(n)

def r_wire_length(i):
  n = 3 * i - 4
  return(n)

def r_degree(m):
  n = []
  for i in range(len(m)):
    if i == 0:
      n.append(2)
    elif i == 1:
      n.append(10/3)
    else:
      j = m[0] * 4 + (m[i-1] - m[0]) * 5 + (m[i] - m[i-1]) * 3
      n.append(j/m[i])
  return(n)
```

```python
r_node = []
for i in range(1, 9):
  r_node.append(r_nodes(i))

r_diametr = []
for i in range(len(r_node)):
  r_diametr.append(int(r_distance(r_node[i])))

r_edges = []
for i in range(len(r_node)):
  r_edges.append(r_num_edges(r_node[i]))

r_average = []
for i in range(len(r_node)):
  n = r_wire_length(r_node[i])
  r_average.append(n/r_edges[i])

r_dencity = []
for i in range(len(r_node)):
  n = edges_full(r_node[i])
  r_dencity.append(r_edges[i]/n)

r_valenсy = []
r_valenсy = r_degree(r_node)

ricobit = pd.DataFrame(list(zip(r_node, r_edges, r_average, r_diametr, r_dencity, r_valenсy)),
               columns =['Number of nodes', 'Number of edges', 'Average distance', 'Diametr', 'Dencity', 'Nodes degree'])
```

```python
def s_nodes(i):
  n = i * 2
  return(n)

def s_num_edges(i):
  if i == 2:
    n = 1
  else:
    n = i * 3 / 2
  return(n)

def s_distance(i):
  n = math.ceil(i/4)
  return(n)

def s_average_dist(i):
  if i % 4 == 0:
    j = i / 4
    n = (2 * j ** 2 + 4 * j + 1) / i
  else:
    j = (i - 2) / 4
    n = (2 * j ** 2 + 2 * j - 1) / i
  return(n)

def s_degree(i):
  if i == 2:
    n = 1
  else:
    n = 3
  return(n)
```

```python
s_node = []
for i in range(1, 271):
  s_node.append(s_nodes(i))

s_edges = []
for i in range(len(s_node)):
  s_edges.append(int(s_num_edges(s_node[i])))

s_diametr = []
for i in range(len(s_node)):
  s_diametr.append(s_distance(s_node[i]))

s_average = []
for i in range(len(s_node)):
  s_average.append(s_average_dist(s_node[i]))

s_dencity = []
for i in range(len(s_node)):
  n = edges_full(s_node[i])
  s_dencity.append(s_edges[i]/n)

s_valenсy = []
for i in range(len(s_node)):
  s_valenсy.append(s_degree(s_node[i]))

spidergon = pd.DataFrame(list(zip(s_node, s_edges, s_average, s_diametr, s_dencity, s_valenсy)),
               columns =['Number of nodes', 'Number of edges', 'Average distance', 'Diametr', 'Dencity', 'Nodes degree'])
```

```python
arr=dim2.transpose().to_numpy()
d2_nodes = np.array(arr[0], int)

d2_edges = []
for i in range(len(d2_nodes)):
  if d2_nodes[i] == 3:
    d2_edges.append(3)
  elif d2_nodes[i] == 4:
    d2_edges.append(6)
  else:
    d2_edges.append(d2_nodes[i]*2)

d2_dencity = []
for i in range(len(d2_nodes)):
  n = edges_full(d2_nodes[i])
  d2_dencity.append(d2_edges[i]/n)

d2_valency = []
for i in range(len(d2_nodes)):
  if d2_nodes[i] == 3:
    d2_valency.append(2)
  elif d2_nodes[i] == 4:
    d2_valency.append(3)
  else:
    d2_valency.append(4)
```

```python
arr=dim3.transpose().to_numpy()
d3_nodes = np.array(arr[0], int)

d3_edges = []
for i in range(len(d3_nodes)):
  if d3_nodes[i] == 6:
    d3_edges.append(15)
  else:
    d3_edges.append(d3_nodes[i]*3)

d3_dencity = []
for i in range(len(d3_nodes)):
  n = edges_full(d3_nodes[i])
  d3_dencity.append(d3_edges[i]/n)

d3_valency = []
for i in range(len(d3_nodes)):
  if d3_nodes[i] == 6:
    d3_valency.append(5)
  else:
    d3_valency.append(6)
```
